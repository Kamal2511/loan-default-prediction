### Cleaning the data

```python
# creating the function for checking the percentage of the null value in the data

def null_value(df):
    null_value = data.isnull().sum()
    null_value_percentage = 100*null_value/len(data)
    null_value_table = pd.concat([null_value, null_value_percentage], axis=1)
    
    # rename the columns in null_value_percentage
    
    null_value_table.rename(columns = {0:'null_value', 1:'%_of_null_value'}, inplace = True)
    
    # sorting the % of total value in descending order\
    
    null_value_table = null_value_table[null_value_table.iloc[:,1] != 0].sort_values('%_of_null_value', ascending= False).round(2)
    
    # print the summary 
    
    print('Dataframe has '+str(data.shape[1])+' columns.\n'+'There are '+str(null_value_table.shape[0]) + 'columns that has null values.')
    
    return null_value_table
```

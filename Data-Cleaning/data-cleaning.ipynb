# **Cleaning the Data**

```python
# creating a function for null values percentage 

def null_value_df(df):
    
    null_value = df.isnull().sum()
    null_value_percentage = 100 * null_value / len(df)  # df.shape[0]
    null_value_table = pd.concat([null_value, null_value_percentage] , axis=1)
    
    # rename the columns
    
    null_value_table.rename(columns = {0 : 'Null Values', 1 : '% of Null Values'},inplace=True)
    
    # sorting the % of Total Values in descending order
    
    null_value_table = null_value_table[null_value_table.iloc[:,1] != 0].sort_values('% of Null Values', ascending=False).round(2)
    
    # print the summary
    
    print ("Dataframe has " + str(df.shape[1]) + " columns.\n"+"There are " + str(null_value_table.shape[0]) + " columns that have null values.")
    
    
    return null_value_table
    
    
```

```python
null_value_df(train_df).head(50)
```

Since the percentage of null value is high, deleting the rows will get data leakage

```python
# filling the null values with the mean

train_df.fillna(train_df.mean, inplace=True)
```

```python
l = []
for i in train_df.select_dtypes(include=['object']).columns:
    
    # Checking columns of object data type
    
    l.append(i)
    
for i in l:
    # deleting the columns with data type objects 
    
    train_df.drop(labels=i, axis=1, inplace=True) 
```
